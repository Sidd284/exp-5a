```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the image
image = cv2.imread('/content/images.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded correctly
if image is None:
    print("Error: Image not found. Please check the file path.")
else:
    # Step 2: Apply the Sobel operator to compute the gradients
    sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in x-direction
    sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in y-direction

    # Step 3: Compute the magnitude of the gradient
    gradient_magnitude = cv2.magnitude(sobel_x, sobel_y)

    # Step 4: Normalize the magnitude to be in the range [0, 255]
    gradient_magnitude = np.uint8(np.clip(gradient_magnitude / np.max(gradient_magnitude) * 255, 0, 255))

    # Step 5: Display the original and edge-detected images
    plt.figure(figsize=(10, 5))

    # Plot original image
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Plot Sobel edge-detected image
    plt.subplot(1, 2, 2)
    plt.imshow(gradient_magnitude, cmap='gray')
    plt.title('Sobel Edge Detection')
    plt.axis('off')

    plt.show()
```
